```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras import models
from keras import layers 
from keras.applications import VGG16
from keras.applications import VGG19
from keras.applications import ResNet50
from keras import optimizers
from keras.layers.core import Flatten, Dense, Dropout, Lambda
from keras.layers import Conv2D, MaxPool2D,BatchNormalization, Activation
from keras.models import Sequential

np.random.seed(0)

os.environ['OMP_NUM_THREADS'] = '1'
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'
tf.__version__

#!unzip Covid_Data_GradientCrescent.zip
```

#### Load Image Data

```python
DATA_LIST = os.listdir('two/train')
DATASET_PATH  = 'two/train'
TEST_DIR =  'two/test'
IMAGE_SIZE    = (224, 224)
NUM_CLASSES   = len(DATA_LIST)
BATCH_SIZE    = 10 
NUM_EPOCHS    = 40
LEARNING_RATE = 0.005 
```

#### Generate Training and Validation Batches

```python
train_datagen = ImageDataGenerator(rescale=1./255,rotation_range=50,featurewise_center = True,
                                   featurewise_std_normalization = True,width_shift_range=0.2,
                                   height_shift_range=0.2,shear_range=0.25,zoom_range=0.1,
                                   zca_whitening = True,channel_shift_range = 20,
                                   horizontal_flip = True,vertical_flip = True,
                                   validation_split = 0.2,fill_mode='constant')




train_batches = train_datagen.flow_from_directory(DATASET_PATH,target_size=IMAGE_SIZE,
                                                  shuffle=True,batch_size=BATCH_SIZE,
                                                  subset = "training",seed=42,
                                                  class_mode="binary")

valid_batches = train_datagen.flow_from_directory(DATASET_PATH,target_size=IMAGE_SIZE,
                                                  shuffle=True,batch_size=BATCH_SIZE,
                                                  subset = "validation",seed=42,
                                                  class_mode="binary")



```

#### Baselines


We decided to run 3 preptrained models with default hyper parameters for a baseline result. We ran VGG16, VGG19 and ResNet50. Among the three of them VGG16 and VGG19 had a good traning and validation accuracy in the range of 92% to 96%. We decided to go ahead with the VGG16 as it has roughly 26% lesser parameters than the VGG19 and hence much less complex even though it provides comparable results.
We also  tried creating my own CNN model but that did not well at all (Validation accuracy ~ 50%). We suspect it could be beacuse of the data limitation or other hyperparameters.

### Own Model Baseline


```python
own = Sequential()
own.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu',input_shape=(224,224,3)))
own.add(MaxPool2D(pool_size=(2,2)))
own.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
own.add(MaxPool2D(pool_size=(2,2)))
own.add(Flatten())
own.add(Dense(units = 16, activation = 'relu'))
own.add(Dropout(rate=  0.25))
own.add(Dense(units=1, activation='sigmoid'))

own.summary()
```

```python
STEP_SIZE_TRAIN=train_batches.n//train_batches.batch_size
STEP_SIZE_VALID=valid_batches.n//valid_batches.batch_size

own.compile(optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),loss='binary_crossentropy',metrics=["accuracy"])

result_own = own.fit_generator(train_batches,steps_per_epoch=STEP_SIZE_TRAIN,
                             epochs=NUM_EPOCHS,validation_data=valid_batches,
                             validation_steps=STEP_SIZE_VALID)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25, 7))
plt.subplot(121)
plt.title("Accuracy")
plt.plot(result_own.history['accuracy'], label='Train accuracy')
plt.plot(result_own.history['val_accuracy'], label = 'validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.4, 1])
plt.legend(loc='lower right')
plt.grid(True)

plt.figure(figsize=(25, 7))
plt.subplot(122)
plt.title("Loss")
plt.plot(result_own.history['loss'], label='Train loss')
plt.plot(result_own.history['val_loss'], label = 'validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim(0.2,1)
plt.legend(loc='upper right')
plt.grid(True)
```

#### Tuning model

In this Part, we decided to tune hyperparameters using VGG16 as the base feature extractor. The parameters that we decided to tune are the learning rate, epochs, Activation fucntion in the Dense Layer and Dropout for the Dense layer. we decided to use GridsearchCV to do the same. we decided to use early stopping to optimize the process and make it faster.

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
import pandas as pd
from sklearn.model_selection import ParameterGrid
from keras.callbacks import EarlyStopping, ModelCheckpoint
es = EarlyStopping(monitor='loss', mode='min', verbose=1, patience= 10)
mc = ModelCheckpoint('best_model.h5', monitor='loss', mode='min', verbose=0, save_best_only=True)


def build_classifier(activation,dropout):
  conv_base = VGG16(weights="imagenet", include_top=False, input_shape=(224, 224, 3))
  conv_base.trainable = False
  classifier = Sequential()
  classifier.add(conv_base)
  classifier.add(Flatten())
  classifier.add(Dense(units = 256, activation = activation))
  classifier.add(Dropout(dropout))
  classifier.add(Dense(units=1, activation='sigmoid'))
  classifier.compile(optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),loss='binary_crossentropy',metrics=["accuracy"])
  return classifier

param = {'activation' : ['sigmoid','relu'],
        'dropout' : [0.15,0.20,0.25],
        'epochs' : [40]}

clf = KerasClassifier(build_classifier)
X_train, y_train = train_batches.next()
X_val, y_val = valid_batches.next()
print(train_batches.class_indices)
print(valid_batches.class_indices)

grid = GridSearchCV(estimator=clf, param_grid=param, scoring='accuracy',verbose=1,cv=3)
grid.fit(X_train,y_train,callbacks=[es,mc])


```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
def build_estimator(learning_rate):
  base = VGG16(weights="imagenet", include_top=False, input_shape=(224, 224, 3))
  base.trainable = False
  estimator = Sequential()
  estimator.add(base)
  estimator.add(Flatten())
  estimator.add(Dense(units = 256, activation = 'relu'))
  estimator.add(Dropout(rate = 0.2))
  estimator.add(Dense(units=1, activation='sigmoid'))
  estimator.compile(optimizer=optimizers.Adam(learning_rate=learning_rate),loss='binary_crossentropy',metrics=["accuracy"])
  return estimator


param2 = {'learning_rate' : [0.001,0.005,0.01,0.05,0.1],
         'epochs' : [40,80,120]}

est = KerasClassifier(build_estimator)
X_train, y_train = train_batches.next()
X_val, y_val = valid_batches.next()
print(train_batches.class_indices)
print(valid_batches.class_indices)

grid2 = GridSearchCV(estimator=est, param_grid=param2, scoring='accuracy',verbose=1,cv=3)
grid2.fit(X_train,y_train,callbacks=[es,mc])
```

```python
print(grid2.best_params_)
print(grid2.best_score_)
```

After running GridSeach for learning_rate, epochs, activation and dropout I found the following values to give the best result. we now retrain the VGG based model with the new hyperparameters

```python
model = Sequential()
model_base = VGG16(weights="imagenet", include_top=False, input_shape=(224, 224, 3))
model_base.trainable = False
model.add(model_base)
model.add(Flatten())
model.add(Dense(units = 256, activation = 'relu'))
model.add(Dropout(rate=0.20))
model.add(Dense(units=1, activation='sigmoid'))


model.compile(optimizer=optimizers.Adam(learning_rate=0.001),loss='binary_crossentropy',metrics=["accuracy"])

```

```python
model.summary()
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
STEP_SIZE_TRAIN=train_batches.n//train_batches.batch_size
STEP_SIZE_VALID=valid_batches.n//valid_batches.batch_size

wt = ModelCheckpoint('wt_1', monitor='accuracy', mode='max', verbose=0, save_best_only=True)

fin_mod = model.fit_generator(train_batches,steps_per_epoch=STEP_SIZE_TRAIN,
                             epochs=120,validation_data=valid_batches,
                             validation_steps=STEP_SIZE_VALID,callbacks=[wt])
```

```python
plt.figure(figsize=(25, 7))
plt.subplot(121)
plt.title("Tuned VGG16 Accuracy")
plt.plot(fin_mod.history['accuracy'], label='Train accuracy')
plt.plot(fin_mod.history['val_accuracy'], label = 'validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.6, 1.1])
plt.legend(loc='lower right')
plt.grid(True)

plt.figure(figsize=(25, 7))
plt.subplot(122)
plt.title("Tuned VGG16 Loss")
plt.plot(fin_mod.history['loss'], label='Train loss')
plt.plot(fin_mod.history['val_loss'], label = 'validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim(0,2)
plt.legend(loc='upper right')
plt.grid(True)
```

#### Plot Test Results

```python
import matplotlib.image as mpimg

#model.load_weights('best_model.h5')

test_datagen = ImageDataGenerator(rescale=1. / 255)
eval_generator = test_datagen.flow_from_directory(TEST_DIR,target_size=IMAGE_SIZE,
                                                  batch_size=1,shuffle=False,seed=42,class_mode="binary")
eval_generator.reset()
pred = model.predict_generator(eval_generator,18,verbose=1)


x = model.evaluate_generator(eval_generator,steps = np.ceil(len(eval_generator)),
                           use_multiprocessing = True,verbose = 1,workers=1)
print('Test loss:' , x[0])
print('Test accuracy:',x[1])



for index, probability in enumerate(pred):
    image_path = TEST_DIR + "/" +eval_generator.filenames[index]
    image = mpimg.imread(image_path)
    if image.ndim < 3:
        image = np.reshape(image,(image.shape[0],image.shape[1],1))
        image = np.concatenate([image, image, image], 2)
#         print(image.shape)

    pixels = np.array(image)
    plt.imshow(pixels)
    
    print(eval_generator.filenames[index])
    if probability > 0.5:
        plt.title("%.2f" % (probability[0]*100) + "% Normal")
    else:
        plt.title("%.2f" % ((1-probability[0])*100) + "% COVID19 Pneumonia")
    plt.show()
```

## Lower dimension plot


```python
from sklearn.manifold import TSNE

intermediate_layer_model = models.Model(inputs=model.input,
                                        outputs=model.get_layer('dense_1').output)

tsne_data_generator = test_datagen.flow_from_directory(DATASET_PATH,target_size=IMAGE_SIZE,
                                                  batch_size=1,shuffle=False,seed=42,class_mode="binary")

logits = intermediate_layer_model.predict_generator(tsne_data_generator,verbose=1,steps = 130)

tsne = TSNE(n_components=2, init = "random", random_state = 44) 
print(logits.shape)
tsne_obj = tsne.fit_transform(logits)
```

```python
x1=[]
y1=[]

x2=[]
y2=[]

for i in range(tsne_obj.shape[0]):
    if(tsne_data_generator.labels[i]==0):
        x1.append(tsne_obj[i,0])
        y1.append(tsne_obj[i,1])
    else:
        x2.append(tsne_obj[i,0])
        y2.append(tsne_obj[i,1])

plt.scatter(x1,y1,c='blue',label = 'COVID19')
plt.scatter(x2,y2,c='red',label = 'Normal')


plt.legend()
```
